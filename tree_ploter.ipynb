```python
from Bio import Phylo
import matplotlib.pyplot as plt

label_map = {
    "AY278741.1": "SARS-CoV",
    "MG772933.1": "Bat-SL-CoVZC45",
    "JX869059.2": "CoV-2c-EM2012",
    "MN988668.1": "WUH01",
    "MN988669.1": "WUH02"
}


tree = Phylo.read("all_sequences_aligned.fasta.treefile", "newick")


for clade in tree.find_clades():
    if clade.name in label_map:
        clade.name = label_map[clade.name]


fig = plt.figure(figsize=(10, 8), facecolor='white')
ax = fig.add_subplot(1, 1, 1)


Phylo.draw(
    tree,
    axes=ax,
    do_show=False,
    label_func=lambda x: x.name,
    branch_labels=lambda c: None,  
)


for text in ax.get_yticklabels():
    text.set_fontsize(10)
    text.set_color("darkblue")

ax.set_title("Phylogenetic Tree of Coronavirus Strains", fontsize=14, fontweight='bold')
ax.set_facecolor("#f7f7f7")
plt.tight_layout()
plt.savefig("tree_labeled_pretty.png", dpi=300)
plt.show()
```
